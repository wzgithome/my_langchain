### 自定义工具

```python
import os
import dotenv
from Cython.Includes.libcpp.vector import vector
from langchain_core.tools import create_retriever_tool
from langchain_tavily import TavilySearch

dotenv.load_dotenv(dotenv_path='../.env')

os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY", "")

search=TavilySearch(max_result=1)

res=search.invoke("今天上海的天气怎么样")
print(res)





```

### 定义Retriever

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

import os
import dotenv
dotenv.load_dotenv(dotenv_path='../.env')
# 初始化模型
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

os.environ["DASHSCOPE_API_KEY"] = os.getenv("OPENAI_API_KEY2")
embedding_model = DashScopeEmbeddings(model="text-embedding-v3")

loader=WebBaseLoader("https://www.sohu.com/a/899930504_799260")
docs=loader.load()

splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

documents=splitter.split_documents(docs)

# 向量化 得到向量数据库对象
vector=Chroma.from_documents(documents,embedding_model)

# 创建检索器
retriever=vector.as_retriever()


# 检测检索结果
print(retriever.invoke("猫的特征")[0].page_content)

```

### 创建工具、工具集

```python
from langchain_core.tools.retriever import create_retriever_tool

# 创建一个工具来检索文档
retriever_tool=create_retriever_tool(
    retriever=retriever,
    name='wiki_search',
    description='百科全书'
)
# 创建工具集
tools=[search,retriever_tool]
```

### 语言模型调用工具

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

llm=ChatOpenAI(
    model='qwen3.5-plus',
    temperature=0.5
)
# 模型绑定工具
model_with_tools=llm.bind_tools(tools)

# 输入自动调用工具
message=[HumanMessage(content="今天上海的天气怎么样")]
res=model_with_tools.invoke(message)
print(f"ContentString:{res}")
print('-------')
print(res.tool_calls)
```
