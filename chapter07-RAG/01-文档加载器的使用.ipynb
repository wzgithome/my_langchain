## 文档加载器

* txt文档：TextLoader
* pdf文档：PyPDFLoader
* csv文档：CSVLoader
* json文档：JSONLoader
* html文档：UnstructuredHTMLLoader
* md文档：UnstructuredMarkdownLoader
* 文件目录：DirectoryLoader

<font color='pink'>创建好XxxxLoader的实例以后，都需要调用load()，在内存中返回
一个list[Document]</font>

LangChain 实现流程
1. 创建一个Loader的实例
2. 调用load()方法，返回一个list[Document]
3. 文档分割器
4. 文本嵌入模型
5. 向量数据库
6. 创建一个RetrievalQA

```python
# 加载Text文件utf-8格式的文件
from langchain_community.document_loaders import TextLoader

# 指明txt文档的路径
file_path='./asset/load/01-langchain-utf-8.txt'

# 创建一个TextLoader的案例
text=TextLoader(
    file_path=file_path,
    # 需要指导文件的编码格式
    encoding='utf-8'
)

# 调用load()，返回一个list[Document]
docs=text.load()
print(docs,'\n')
# 显示Document对象的元数据
print(docs[0].metadata,'\n')
# 显示文档中的内容信息
print(docs[0].page_content)
```

```python
# 加载pdf文件
from langchain_community.document_loaders import PyPDFLoader


pdf=PyPDFLoader(
    # 可以加载本地的，也可以加载网络中的资源
    file_path='./asset/load/02-load.pdf'
)

docs=pdf.load()
print(docs,'\n')
print(docs[0].metadata)
```

```python
# 加载CSV文档
from langchain_community.document_loaders import CSVLoader

csv=CSVLoader(
    file_path='./asset/load/03-load.csv'
)
docs=csv.load()

print(len(docs))
print(docs)    # 有多个metadata
print(docs[0].metadata,'\n')
for doc in docs:
    print(doc)

```

```python
# 举例1:加载JSON文件
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.',  #表示加载所有的字段
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
print(docs)
```

```python
# 举例2:加载JSON文件里面的某个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.messages[].content',  #表示加载messages[]的所有的content字段
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
for doc in docs:
    print(doc.page_content)
```

```python
# 举例3:加载JSON文件里面的某个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[].content',  #表示加载data.items[].content
    text_content=False, #将加载的json对象转换为json字符串
)
docs=json.load()
print(docs,'\n')

for doc in docs:
    print(doc.page_content)
```

```python
# 举例3:第二种方式
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[]',  #表示加载data.items[].content
    content_key='.content',
    is_content_key_jq_parsable=True # 用jq解析content_key
)
docs=json.load()


for doc in docs:
    print(doc.page_content)
```

```python
# 举例4:获取2个字段
from langchain_community.document_loaders import JSONLoader

json=JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[]',  #表示加载data.items[].content
    content_key='.content+"\\n\\n"+.title',
    is_content_key_jq_parsable=True # 用jq解析content_key
)
docs=json.load()


for doc in docs:
    print(doc.page_content)
```

```python
# 举例1：加载HTML文件
from langchain_community.document_loaders import UnstructuredHTMLLoader

'''
    定义UnstructuredHTMLLoader对象
    strategy：
        “fast”解析加载html文件速度是比较快（但可能丢失部分结构或元数据）
        “hi_res”:（高分辨率解析）解析精准（速度慢一些）
        “ocr_only”:强制使用ocr提起文本，仅仅适用于图像（对HTML无效）
    mode：
        one of{paged，element，single}
        elements：按语义元素（标题、段落、列表、表格等）拆分成多个独立的小文档
'''
html=UnstructuredHTMLLoader(
    file_path="asset/load/05-load.html",
    mode='elements',
    strategy='fast'
)
# 加载
docs=html.load()
print(len(docs))

for doc in docs:
    print(doc)

```

```python
# 举例：加载markdown文档
from langchain_community.document_loaders import UnstructuredMarkdownLoader
from pprint import pprint

mark=UnstructuredMarkdownLoader(
    file_path='asset/load/06-load.md',
    strategy='fast'
)
# 加载
docs=mark.load()
print(len(docs))
# 打印
for doc in docs:
    pprint(doc)
```

```python
# 加载指定目录
from langchain_community.document_loaders import DirectoryLoader,PythonLoader
from pprint import pprint


directory=DirectoryLoader(
    path='./asset/load',
    glob='*.py',
    use_multithreading=True,
    show_progress=True,
    loader_cls=PythonLoader
)

# 加载
docs=directory.load()
print(len(docs))
# 打印
# docs
for doc in docs:
    pprint(docs)
```
