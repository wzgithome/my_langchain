```python
# 举例1:文档向量化
from langchain_community.embeddings import DashScopeEmbeddings
import os
import dotenv
# 1. 加载环境变量
dotenv.load_dotenv(dotenv_path='../.env')

# 2. 设置你的 DashScope API Key
# 注意：这里的变量名通常是 DASHSCOPE_API_KEY
os.environ["DASHSCOPE_API_KEY"] = os.getenv("OPENAI_API_KEY2")

embedding = DashScopeEmbeddings(model="text-embedding-v3")

texts=[
    'Hi there',
    'On,hello',
    '你的名字',
    '林黛玉到把垂杨柳',
    '武大郎拳打镇关西'
]

embeddings=embedding.embed_documents(texts)

for i in range(len(texts)):
    print(f'{texts[i]}:{embeddings[i][:3]}')


```

```python
# 举例2
import os
import dotenv
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_community.document_loaders import CSVLoader

load=CSVLoader('./asset/load/03-load.csv')
docs=load.load_and_split()

dotenv.load_dotenv(dotenv_path='../.env', encoding='utf-8')

# 2. 设置你的 DashScope API Key
# 注意：这里的变量名通常是 DASHSCOPE_API_KEY
os.environ["DASHSCOPE_API_KEY"] = os.getenv("OPENAI_API_KEY2")

# 3. 创建 Embedding 实例，指定你想要的模型
# 如果想使用 qwen3-vl-embedding，直接在这里指定 model 参数
embedding = DashScopeEmbeddings(model="text-embedding-v3")


embedding_docs=embedding.embed_documents([doc.page_content for doc in docs])
print(len(embedding_docs))
print(len(embedding_docs[0]))
print(embedding_docs[0][:10])
```
