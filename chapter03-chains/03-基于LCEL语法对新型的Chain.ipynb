## creat_sql_query_chain的使用

```python
from langchain_classic.chains.sql_database.query import create_sql_query_chain
from langchain_openai import ChatOpenAI
from langchain_community.utilities import SQLDatabase
import os
import dotenv
dotenv.load_dotenv(dotenv_path='../.env')
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")
# 1. 连接数据库 (例如 Mysql)
db = SQLDatabase.from_uri(
    "mysql+pymysql://root:root@localhost:3306/llm",

)

# 2. 初始化模型
llm = ChatOpenAI(
    model='tongyi-xiaomi-analysis-pro',
    temperature=0
)

# 3. 创建 SQL 查询链
# 该链会自动读取 db 的表结构信息并填充到 Prompt 中
chain = create_sql_query_chain(llm, db)

# 4. 执行链
# query 参数是自然语言问题(默认只能写question)
response = chain.invoke({"question": "家庭住址是西安的员工信息"})

print(response)
# 输出: SELECT count(*) FROM employees WHERE department = 'Finance';
```
