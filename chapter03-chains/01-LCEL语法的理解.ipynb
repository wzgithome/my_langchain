## 关于LCEL语法使用

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI
import os
import dotenv
dotenv.load_dotenv()

dotenv.load_dotenv(dotenv_path='../.env')
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY2")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL2")

# 1. 定义数据结构 (使用 Pydantic) 模型输出的键值
class QuestionAnswer(BaseModel):
    setup: str = Field(description="问题部分")
    punchline: str = Field(description="结果部分")

# 2. 初始化解析器，并将结构传入
parser = JsonOutputParser(pydantic_object=QuestionAnswer)

prompt=PromptTemplate.from_template(
    template="{format_instructions}\n 问题是{query}"
)
prompt = prompt.partial(format_instructions=parser.get_format_instructions())



model =  ChatOpenAI(
    model='qwen3.5-plus',
    temperature=0.6
)

# model如果是非对话式返回的是str，如果是对话式返回的是AiMessage
chain=prompt | model | parser
print(chain.invoke({"query": "什么是C语言"}))

```
