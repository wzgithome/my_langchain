```python
from langchain.chat_models import init_chat_model

import dotenv
import os

"""
    temperature:
            0.0~0.3：需要一致性、准确性的任务
            0.5~0.7：平衡创造性和一致性
            0.8~1.5：创造性任务
            1.5~2.0：高度创造性
"""
dotenv.load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY2")
OPENAI_BASE_URL = os.getenv("OPENAI_BASE_URL2")
init_model = init_chat_model(
    model_provider='openai',
    model='qwen3.7-max',
    api_key=OPENAI_API_KEY,
    base_url=OPENAI_BASE_URL,
    # 开启深度思考
    extra_body={
        "enable_thinking": True,
    },
)

res = init_model.invoke("你是什么模型？")
# 查看思考过程
print("=== 思考过程 ===")

print(res.response_metadata)
print("\n=== 最终回答 ===")
print(res.content)
```

### 部分变量预填充：partial()

```python
from langchain_core.prompts import ChatPromptTemplate
from llm.my_llm import model

# 原始模版
template = ChatPromptTemplate.from_messages([
    ('system', '你是{department},目标用户是{role}'),
    ('user', '{task}')
])
# 部分填充
it_template = template.partial(
    department='IT 部门',
    role='技术支持'
)
# 销售部门
sales_template = template.partial(
    department='销售部门',
    role='销售顾问'
)
sales=sales_template.invoke({'task':'为什么每年年底汽车会促销'})

res=model.invoke(sales)
print(res.content)
```
