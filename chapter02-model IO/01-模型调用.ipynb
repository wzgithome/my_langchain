```python
from langchain_openai import ChatOpenAI
import os
import dotenv
dotenv.load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY5")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL5")
llm = ChatOpenAI(
    model='mimo-v2.5-pro',
    # model='GLM-4.7-Flash',
)
res=llm.invoke('大模型是什么')
# res=llm.invoke('根据IP是192.168.0.158查询所在位置')
print(res.content)
```

```python
from langchain_community.chat_models import ChatTongyi

# 读取.env配置文件中的信息。相关的环境变量以.env文件中的优先
import dotenv
dotenv.load_dotenv(override=True)
llm_tongyi=ChatTongyi(
    model='qwen3.7-max'
)

response=llm_tongyi.invoke('解释一下你自己')
print(response.content)
```

```python
from langchain.chat_models import init_chat_model
import dotenv
import os
"""
    temperature:
            0.0~0.3：需要一致性、准确性的任务
            0.5~0.7：平衡创造性和一致性
            0.8~1.5：创造性任务
            1.5~2.0：高度创造性
"""
dotenv.load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY2")
OPENAI_BASE_URL = os.getenv("OPENAI_BASE_URL2")
init_model=init_chat_model(
    model_provider='openai',
    model='qwen3.7-plus',
    api_key=OPENAI_API_KEY,
    base_url=OPENAI_BASE_URL,
)


res=init_model.invoke("你是什么模型？")
print(res)
```
